```
def savegrid(ims, rows=None, cols=None, fill=True, showax=False):
    if rows is None != cols is None:
        raise ValueError("Set either both rows and cols or neither.")

    if rows is None:
        rows = 1
        cols = len(ims)

    gridspec_kw = {'wspace': 0, 'hspace': 0} if fill else {}
    fig,axarr = plt.subplots(rows, cols, gridspec_kw=gridspec_kw, figsize=(10,10))

    if fill:
        bleed = 0
        fig.subplots_adjust(left=bleed, bottom=bleed, right=(1 - bleed), top=(1 - bleed))

    for ax,im in zip(axarr.ravel(), ims):
        ax.imshow(im)
        if not showax:
            ax.set_axis_off()

    kwargs = {'pad_inches': .01} if fill else {}
    fig.savefig('image.png', **kwargs)
```

```
# Read Bakground images 
Background_folder = 'Background/'
bg_images=[]
for i in random.sample(glob(Background_folder+'*.jpg'),6):
  bg_images.append(Image.open(i))

savegrid(bg_images)
```

```
# Read Foreground images 
Foreground_folder = 'Foreground/'
fg_images=[]
for i in random.sample(glob(Foreground_folder+'*.png'),6):
  fg_images.append(Image.open(i))

savegrid(fg_images)
```
